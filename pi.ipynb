```python
#Answer of 1
# create and import module
from math import factorial

def e_approx(n):
    e = 0
    # Loop through till n
    for i in range(n + 1):
        # add it to e
        e += 1 / factorial(i)
    # return the value of e
    return e


# Testing
if __name__ == '__main__':
    for n in range(10,100):
        e = e_approx(n)
        print("n=", n, " ==> e=", e, end=" ")
        
```

```python
#Answer of 2
def squares(s):
    return [round(n) for n in s for x in range(1,n+1) if x*x==n]


# Testing
print(squares( [49,8,2,1,102]))
print(squares( [500,30]))
```
